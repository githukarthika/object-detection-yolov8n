```python
from google.colab import drive
drive.mount('/content/drive')
import os
image_path = '/content/drive/MyDrive/photo-1611267254323-4db7b39c732c.jpg'
print("Image uploaded from drive:", image_path)
```

```python

model = YOLO('yolov8n.pt')
```

```python
results = model.predict(source=image_path,save=True)
from IPython .display import Image
Image(filename='runs/detect/predict/' + os.path.basename(image_path))
```

```python
for r in results:
  boxes = r.boxes
  for box in boxes:
    cls_id=int(box.cls)
    cls_name=model.names[cls_id]
    confidence=float(box.conf)
    coords=box.xyxy.tolist()
    print(f"class: {cls_name}, confidence: {confidence:.2f}, Box coordinates: {coords}")
```
